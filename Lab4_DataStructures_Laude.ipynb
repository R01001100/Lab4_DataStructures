```python
# Jaccard Similarity Algorithm


def calculate_jaccard_similarity(set1, set2, user_name):
    """
    Calculate Jaccard Similarity between two sets
    J(A,B) = |A ∩ B| / |A ∪ B|
    Returns percentage value
    """
    # Calculate intersection and union
    intersection = set1.intersection(set2)
    union = set1.union(set2)

    # Calculate Jaccard Index
    if len(union) == 0:  # Avoid division by zero
        similarity = 0
    else:
        similarity = (len(intersection) / len(union)) * 100

    print(f"\n{user_name} Jaccard Analysis")
    print(f"  Intersection (artists in common): {intersection}")
    print(f"  Size of Intersection: {len(intersection)}")
    print(f"  Union (all unique artists): {union}")
    print(f"  Size of Union: {len(union)}")
    print(f"  Jaccard Similarity: {similarity:.2f}%")

    return similarity, intersection, union


# Calculate for User A
jaccard_a, intersect_a, union_a = calculate_jaccard_similarity(
    user_a, festival, "User A"
)

# Calculate for User B
jaccard_b, intersect_b, union_b = calculate_jaccard_similarity(
    user_b, festival, "User B"
)
```

```python
print("DEALBREAKERS ANALYSIS")

# Find User A's dealbreakers (artists they listen to that are NOT in festival)
dealbreakers_a = user_a.difference(festival)

print(f"\nUser A's Listening Profile: {user_a}")
print(f"Expanded Festival Lineup: {festival}")
print(f"\nUser A's DEALBREAKERS (artists NOT performing at festival):")
if dealbreakers_a:
    for artist in sorted(dealbreakers_a):
        print(f"  ✗ {artist}")
    print(f"\nTotal dealbreakers: {len(dealbreakers_a)}")
    print(
        f"Dealbreaker percentage: {(len(dealbreakers_a)/len(user_a))*100:.2f}% of User A's library"
    )
else:
    print("  ✓ None! All of User A's artists are performing at the festival.")
```

```python
print("FESTIVAL RECOMMENDATION ANALYSIS")

# Determine which user is a better match for the festival
if jaccard_a > jaccard_b:
    better_match = "User A"
    better_score = jaccard_a
    worse_score = jaccard_b
elif jaccard_b > jaccard_a:
    better_match = "User B"
    better_score = jaccard_b
    worse_score = jaccard_a
else:
    better_match = "Both users equally"
    better_score = jaccard_a

print(f"\nRECOMMENDATION RESULTS:")
print(f"• User A Similarity: {jaccard_a:.2f}%")
print(f"• User B Similarity: {jaccard_b:.2f}%")
print(f"\n• BEST MATCH FOR FESTIVAL: {better_match} ({better_score:.2f}%)")
print(f"• Difference: {abs(better_score - worse_score):.2f}%")

# Find artists that User B enjoys that User A might like (for cross-recommendation)
user_b_exclusive = user_b.difference(user_a)
artists_in_festival = user_b_exclusive.intersection(festival)

print(f"\n• Artists User B enjoys that are in festival (recommend to User A):")
if artists_in_festival:
    for artist in sorted(artists_in_festival):
        print(f"  → {artist}")
```

```python
# Initialize Base Burst

from collections import deque, Counter

# Initialize deque with strict memory limit of 5
alert_buffer = deque(maxlen=5)

# Simulate burst of incoming alerts
burst = [
    (CONTROL_NUM, "WEATHER"),  # (4, "WEATHER")
    (CONTROL_NUM + 1, "TRAFFIC"),  # (5, "TRAFFIC")
    (CONTROL_NUM + 2, "WEATHER"),  # (6, "WEATHER")
    (CONTROL_NUM + 3, "WEATHER"),  # (7, "WEATHER")
    (CONTROL_NUM + 4, LOCAL_HAZARD),  # (8, "TYPHOON")
    (CONTROL_NUM + 5, "WEATHER"),  # (9, "WEATHER")
    (CONTROL_NUM + 6, "TRAFFIC"),  # (10, "TRAFFIC")
]

print("INITIAL BURST")
for alert in burst:
    print(f"  {alert}")
print(f"Total initial alerts: {len(burst)}")
```

```python
# Data Expansion Phase
# Add exactly (CONTROL_NUM + 2) custom alerts
# CONTROL_NUM = 4, so 4 + 2 = 6 custom alerts needed

print(f"Need to add {CONTROL_NUM + 2} custom alerts (CONTROL_NUM + 2 = 4 + 2 = 6)")

# Find the last timestamp to continue sequentially
last_timestamp = burst[-1][0]

custom_alerts = [
    (last_timestamp + 1, "EARTHQUAKE"),  # (11, "EARTHQUAKE")
    (last_timestamp + 2, "WEATHER"),  # (12, "WEATHER")
    (last_timestamp + 3, "TRAFFIC"),  # (13, "TRAFFIC")
    (last_timestamp + 4, LOCAL_HAZARD),  # (14, "TYPHOON")
    (last_timestamp + 5, "FLOOD"),  # (15, "FLOOD")
    (last_timestamp + 6, "WEATHER"),  # (16, "WEATHER")
]

print(
    f"\nCustom alerts to add (timestamps {last_timestamp + 1} to {last_timestamp + 6}):"
)
for alert in custom_alerts:
    print(f"  {alert}")

# Extend the burst list
burst.extend(custom_alerts)

print("\n=== DATA EXPANSION PHASE ===")
print(f"Added {CONTROL_NUM + 2} custom alerts (timestamps 11-16):")
for alert in custom_alerts:
    print(f"  {alert}")
print(f"\nTotal expanded burst length: {len(burst)}")
print("All alerts in burst (chronological order):")
for i, alert in enumerate(burst, 1):
    print(f"  {i:2d}. {alert}")
```

```python
# Initialize tracking variables
rejected_alerts = []
accepted_alerts = []

# Track specific alerts for assessment questions
weather_at_control_plus_3_appended = False  # For timestamp CONTROL_NUM + 3 (7)
local_hazard_appended = False  # For LOCAL_HAZARD (TYPHOON)

print(
    f"Weather at timestamp {CONTROL_NUM + 3} appended: {weather_at_control_plus_3_appended}"
)
print(f"LOCAL_HAZARD ({LOCAL_HAZARD}) appended: {local_hazard_appended}")
```

```python
# Process All Alerts Through Sliding Window Filter

# Process each alert through the sliding window filter
for i, (timestamp, category) in enumerate(burst, 1):
    # Step 1: Extract current categories from deque using list comprehension
    current_categories = [alert[1] for alert in alert_buffer]

    # Step 2: Count frequencies using Counter
    category_counter = Counter(current_categories)

    # Step 3: Apply spam rule
    # If category already appears 2 or more times in current buffer, reject
    if category_counter[category] >= 2:
        status = "REJECTED (spam)"
        rejected_alerts.append((timestamp, category))
        decision = "[X] REJECTED"
    else:
        # Accept the alert and append to deque
        alert_buffer.append((timestamp, category))
        status = "ACCEPTED"
        accepted_alerts.append((timestamp, category))
        decision = "[/] ACCEPTED"

        # Track specific alerts for assessment
        if timestamp == CONTROL_NUM + 3:  # Timestamp 7
            weather_at_control_plus_3_appended = True
        if category == LOCAL_HAZARD:  # TYPHOON
            local_hazard_appended = True

    # Print detailed processing log
    print(f"\nAlert {i}: ({timestamp}, '{category}')")
    print(f"  Current buffer categories: {current_categories}")
    print(f"  Category counts: {dict(category_counter)}")
    print(f"  '{category}' count in buffer: {category_counter[category]}")
    print(f"  Decision: {decision}")
    print(f"  Buffer after: {list(alert_buffer)}")

print("\nPROCESS COMPLETE")
```

```python
# Filtering Summary

print(f"Total alerts processed: {len(burst)}")
print(f"Accepted alerts: {len(accepted_alerts)}")
print(f"Rejected alerts: {len(rejected_alerts)}")

print("\n[/] ACCEPTED Alerts (in chronological order):")
for alert in accepted_alerts:
    print(f"  ✓ {alert}")

print("\n[X] REJECTED Alerts (spam filtered):")
for alert in rejected_alerts:
    print(f"  ✗ {alert}")
```

```python
# Final Deque Analysis and Timestamp Summation

# Get final state of deque
final_buffer = list(alert_buffer)
print(f"Final buffer state (maxlen=5): {final_buffer}")
print(f"Final deque length: {len(final_buffer)}")

# Sum all timestamp values in final deque
timestamps_in_buffer = [timestamp for timestamp, _ in final_buffer]
timestamp_sum = sum(timestamps_in_buffer)

print(f"\nTimestamps in final buffer: {timestamps_in_buffer}")
print(f"Sum of timestamps in final deque: {timestamp_sum}")
```
